```python
import pandas as pd
import os
```

```python
# columns: data_id, h5ad_path, results_path/markers_path, cxg_flag, var_col_arg, cluster_header_arg, species
studies = {"HLCA_Core" : ["/data/ScheuermannGroup/cellxgene/Ajith/cellxgene/hlcacore.h5ad", "/data/pintardde/class_marker_pipeline/results/HLCA_Core_results"]}
path_to_data = "/data/pintardde/combined_marker_pipeline/data"
files = os.listdir(path_to_data)

base_res_path = "/data/pintardde/combined_marker_pipeline/v2_results" # remember this is what I am calling the new, updated results directory   

# ADD NECESSARY FILE PATHS TO SAMPLE SHEET
for file in files:
    if file.endswith(".h5ad"):
        if not("downsampled_" in file) and not("mouse_EC.clean.integrated." in file):
            data_id = file.replace("(edited_var).h5ad", "") if "(edited_var).h5ad" in file else file.replace(".h5ad", "")
            studies[data_id] = [
                str(os.path.join(path_to_data, file)), # h5ad_path
                str(os.path.join(base_res_path, f"{data_id}_results"))]# results_path/markers_path

for k,v in studies.items():
    if k in ['33208946_travaglini_2020_nature_human_10x', 
             '33208946_travaglini_2020_nature_human_smartseq2', 
             '39818203_niethamer_2025_cellstemcell_mouse',
             'HLCA_Core', 'tabula_senis_mouse_10x', 'tabula_senis_mouse_smartseq2']:
                v.append(True)
    else:
        v.append(False)
    
    # add .var params
    if k == 'LungMap_CellRef': v.append("features")
    else: v.append("")
    
    # add cluster_header param
    if k == 'LungMap_CellRef': v.append("celltype_level3_fullname")
    elif k == 'HLCA_Core': v.append("ann_finest_level")
    elif k == "39818203_niethamer_2025_cellstemcell_mouse": v.append("author_cell_type")
    elif k == "31840053_raredon_2019_sciadv_rat_10x_Type.pub.updated" : v.append("Type.pub.updated")
    elif k == "34030460_schupp_2021_circ_mouse" : v.append("cell.type.ident")
    elif k == "UPENN_PedDev_seurat" : v.append("celltype") 
    else: v.append("free_annotation") # for x2 beta-HLCA and x2 tabula muris
    
    # add species params for handling feature labels
    if k in ['39818203_niethamer_2025_cellstemcell_mouse', "tabula_senis_lung_10x", "tabula_senis_lung_smartseq2", "34030460_schupp_2021_circ_mouse"]: v.append("mouse")
    elif k == '31840053_raredon_2019_sciadv_rat_10x_Type.pub.updated': v.append("rat")
    else: v.append("human")
    
    # add chemistry information per dataset
    if k in ["tabula_senis_lung_smartseq2", '33208946_travaglini_2020_nature_human_smartseq2']:
        v.append("SMART-seq2")
    else:
        v.append("10x") # need to double check for Morley and Raredon as of 2:25 PM 06/08/2026
    
    # add endo_labels
    if k == "HLCA_Core" : v.append("EC venous systemic,EC arterial,EC venous pulmonary,Lymphatic EC differentiating,EC aerocyte capillary,EC general capillary,Lymphatic EC mature,Lymphatic EC proliferating")
    elif k == 'LungMap_CellRef': v.append("Capillary 1 cell,Capillary 2 cell,Venous endothelial cell,Arterial endothelial cell,Lymphatic endothelial cell,Systemic venous endothelial cell")
    elif k == "39818203_niethamer_2025_cellstemcell_mouse" : v.append("CAP1,CAP2,Lymphatic_endothelium,Venous_endothelium,Arterial_endothelium")
    elif k == "31840053_raredon_2019_sciadv_rat_10x_Type.pub.updated": v.append("Arterial-Endothelium,Venous-Endothelium,aCap,gCap,Lymphatic-Endothelium,Cycling-Endothelium")
    elif k == "34030460_schupp_2021_circ_mouse": v.append("EC_arterial,EC_capillary_A,EC_capillary_B,EC_venous,EC_lymph,EC_bronchial")
    elif k == "33208946_travaglini_2020_nature_human_10x" : v.append("Capillary Aerocyte,Capillary,Capillary Intermediate 1,Capillary Intermediate 2,Vein,Artery,Lymphatic,Bronchial Vessel 2,Bronchial Vessel 1")
    elif k == "33208946_travaglini_2020_nature_human_smartseq2": v.append("Capillary Aerocyte,Capillary Intermediate 1,Artery,Capillary,Vein,Lymphatic,Bronchial Vessel 1")
    elif k == "tabula_senis_lung_10x" : v.append("Artery,Capillary,Capillary Aerocyte,Lympatic,Vein")
    elif k == "tabula_senis_lung_smartseq2" : v.append("Lympatic,Lymphatic Cell,Artery,Capillary,Capillary Aerocyte,Capillary Type 1 Cell, Capillary Type 2 Cell,Lymphatic Cell,Vein")
    elif k == "UPENN_PedDev_seurat" : v.append("AEC,CAP1,CAP2,CAP1_CAP2,CAP_proliferating,VEC,SEC,LEC")
    
```
