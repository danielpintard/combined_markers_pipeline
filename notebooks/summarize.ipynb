```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import ast
import os

seed = 42
```

# Summarize results in table

```python
all_res_path = "/data/pintardde/combined_marker_pipeline/v2_results/"

# [ ] TODO: format list to reflect which dataset each set of endothelial labels come from for documentation purposes
# [ ] TODO: add in Raredon and Morley endothelial labels as well
endothelial_labels = [
    "Capillary 1 cell", "Capillary 2 cell", "Venous endothelial cell", 'Arterial endothelial cell', 'Lymphatic endothelial cell', 'Systemic venous endothelial cell', # LungMap CellRef endothelial labels
    "EC venous systemic", "EC arterial", "EC venous pulmonary", "Lymphatic EC differentiating", "EC aerocyte capillary", "EC general capillary", "Lymphatic EC mature", "Lymphatic EC proliferating", # HLCA Core endothelial labels
    "CAP1", "CAP2", "Lymphatic_endothelium", "Venous_endothelium", "Arterial_endothelium", # 39818203_niethamer endothelial labels
    "Capillary Aerocyte", "Capillary Intermediate 1", "Artery", "Capillary", "Vein", "Lymphatic," "Bronchial Vessel 1", # travaglini 10x endothelial labels
    "Capillary Aerocyte", "Capillary", "Capillary Intermediate 1", "Capillary Intermediate 2", "Vein", "Artery", "Lymphatic", "Bronchial Vessel 2", "Bronchial Vessel 1", # travaglini SMART-Seq2 endothelial labels
    "Artery", "Capillary", "Capillary Aerocyte", "Lympatic", "Vein", # tabula senis 10x endothelial labels
    "Lympatic", "Lymphatic Cell", "Artery", "Capillary", "Capillary Aerocyte", "Capillary Type 1 Cell", "Capillary Type 2 Cell", "Lymphatic Cell,Vein", # tabula senis SMART-Seq2 endothelial labels
    "Arterial-Endothelium", "Venous-Endothelium", "aCap", "gCap", "Lymphatic-Endothelium", "Cycling-Endothelium", # raredon Type.pub.updated rat endothelial labels
    "EC_arterial", "EC_capillary_A", "EC_capillary_B", "EC_venous", "EC_lymph", "EC_bronchial", # schupp mouse endothelial labels
    "AEC", "CAP1", "CAP2", "CAP1_CAP2", "CAP_proliferating", "VEC", "SEC", "LEC" # Morley celltype endothelial labels
    ]

combined_marker_res = []

for folder in os.listdir(all_res_path):
    if any(x in folder for x in ['ds_', 'downsampled', 'normalized', 'split_by_study', '_annotation_2019', 'test_nods', "UPENN_PedDev_seurat_adult_HLCA_ann_finest_level", "UPENN_PedDev_seurat_adult_LMCR.celltype_level3_results"]):
        continue
    else:
        key = folder.replace("_results", "")
        cmb_mrkr_file = os.path.join(all_res_path, folder, 'tables', 'combined_markers_eval_results.csv')
        df = pd.read_csv(cmb_mrkr_file)
        df = df[df['clusterName'].isin(endothelial_labels)]
        df['study_key'] = key
        df = df[['study_key', 'clusterName', 'clusterSize', 'f_score', 'PPV', 'recall', 'marker_count', 'markers']]
        combined_marker_res.append(df)
    
```

```python
local_marker_res = []

for folder in os.listdir(all_res_path):
    if any(x in folder for x in ['ds_', 'downsampled', 'normalized', 'split_by_study', '_annotation_2019', 'test_nods', 
                                 "UPENN_PedDev_seurat_adult_HLCA_ann_finest_level", "UPENN_PedDev_seurat_adult_LMCR.celltype_level3_results"]):
        continue
    else:
        key = folder.replace("_results", "")
        loca_mrkr_file = os.path.join(all_res_path, folder, 'tables', 'local_marker_eval_on_global_results.csv')
        df = pd.read_csv(loca_mrkr_file)
        df = df[df['clusterName'].isin(endothelial_labels)]
        df['study_key'] = key
        df = df[['study_key', 'clusterName', 'marker_count', 'markers']]
        df = df.rename(columns = {
            "marker_count" : "local_marker_count",
            "markers" : "local_markers"
            })
        local_marker_res.append(df)
```

```python
class_marker_res = []

for folder in os.listdir(all_res_path):
    if any(x in folder for x in ['ds_', 'downsampled', 'normalized', 'split_by_study', '_annotation_2019', 'test_nods', 
                                 "UPENN_PedDev_seurat_adult_HLCA_ann_finest_level", "UPENN_PedDev_seurat_adult_LMCR.celltype_level3_results"]):
        continue
    else:
        key = folder.replace("_results", "")
        class_mrkr_file = os.path.join(all_res_path, folder, 'tables', 'class_and_global_NSF_results.csv')
        df = pd.read_csv(class_mrkr_file)
        df = df[df['clusterName'] == "Endothelial"]
        df['study_key'] = key
        df = df[['study_key', 'clusterName', 'marker_count', 'NSForest_markers']]
        df = df.rename(columns = {
            "marker_count" : "class_marker_count",
            "NSForest_markers" : "class_markers"
            })
        class_marker_res.append(df)
```

```python
all_cmb_mkr_df = pd.concat(combined_marker_res, ignore_index=False)
all_loc_mkr_df = pd.concat(local_marker_res)
all_class_mkr_df = pd.concat(class_marker_res)

res_out_path = "../data/out_data/compiled_NSF_combined_marker_res"
os.makerdirs(res_out_path)

all_cmb_mkr_df.to_csv(os.path.join(res_out_path, "all_combined_markers_eval_06_10_2026.csv"))
all_loc_mkr_df.to_csv(os.path.join(res_out_path, "all_local_markers_eval_06_10_2026.csv"))
all_class_mkr_df.to_csv(os.path.join(res_out_path, "all_class_markers_results_06_10_2026.csv"))
```
