# Summarize results in table

```python
all_res_path = "/data/pintardde/combined_marker_pipeline/results/"

# [ ] TODO: format list to reflect which dataset each set of endothelial labels come from for documentation purposes
# [ ] TODO: add in Raredon and Morley endothelial labels as well
endothelial_labels = ["Capillary 1 cell", "Capillary 2 cell", "Venous endothelial cell", 'Arterial endothelial cell', 'Lymphatic endothelial cell', 'Systemic venous endothelial cell', 
                      "EC venous systemic", "EC arterial", "EC venous pulmonary", "Lymphatic EC differentiating", "EC aerocyte capillary", "EC general capillary", "Lymphatic EC mature", "Lymphatic EC proliferating",
                      "CAP1", "CAP2", "Lymphatic_endothelium", "Venous_endothelium", "Arterial_endothelium", "Capillary Aerocyte", "Capillary Intermediate 1", "Artery", "Capillary", "Vein", "Lymphatic,Bronchial Vessel 1", 
                      "Capillary Aerocyte", "Capillary", "Capillary Intermediate 1", "Capillary Intermediate 2", "Vein", "Artery", "Lymphatic", "Bronchial Vessel 2", "Bronchial Vessel 1", "Artery", "Capillary", "Capillary Aerocyte", 
                      "Lympatic", "Vein", "Lympatic", "Lymphatic Cell", "Artery", "Capillary", "Capillary Aerocyte", "Capillary Type 1 Cell", "Capillary Type 2 Cell", "Lymphatic Cell,Vein", "Arterial-Endothelium", "Venous-Endothelium",
                      "aCap", "gCap", "Lymphatic-Endothelium", "Cycling-Endothelium", "EC_vasc", "EC_Vcam1+", "EC_cap", "EC_lymph", "EC_arterial", "EC_capillary_A", "EC_capillary_B", "EC_venous", "EC_lymph", "EC_bronchial"]

combined_marker_res = []

for folder in os.listdir(all_res_path):
    if any(x in folder for x in ['ds_', 'downsampled', 'normalized', 'split_by_study', '_annotation_2019', 'test_nods']):
        continue
    else:
        key = folder.replace("_results", "")
        file = os.path.join(all_res_path, folder, 'tables', 'combined_markers_eval_results.csv')
        df = pd.read_csv(file)
        df = df[df['clusterName'].isin(endothelial_labels)]
        df['study_key'] = key
        df = df[['study_key', 'clusterName', 'clusterSize', 'f_score', 'precision', 'recall', 'marker_count', 'markers', 'onTarget']]
        combined_marker_res.append(df)
```

```python
all_cmb_mkr_df = pd.concat(combined_marker_res, ignore_index=False)
```
